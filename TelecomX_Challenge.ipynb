```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/challenge2-data-science/refs/heads/main/TelecomX_Data.json'
```

```python
df = pd.read_json(url)
```

```python
df.head()
```

```python
df.info()
```

#ðŸ”§ TransformaÃ§Ã£o

```python
#normalizando todas as colunas de uma Ãºnica vez e concatenando os dados a um Ãºnico dataframe

colunas = ['customer','phone','internet','account']
colunas_normalizadas = []

for col in colunas:
  col_normal = pd.json_normalize(df[col], sep='_')
  colunas_normalizadas.append(col_normal)

dados = pd.concat([
    df[['customerID', 'Churn']],
    *colunas_normalizadas
], axis=1)

dados.head()
```

```python
dados.info()
```

```python
#Verifica dados duplicados
print('NÃºmero de duplicados', dados.duplicated().sum())
```

```python
# Verifica dados nulos
print(dados.isnull().sum())
```

```python
# Verifica NaN ou strings vazias
(dados.apply(lambda x: x.str.strip() == '' if x.dtype == 'object' else False).fillna(False) | dados.isna()).sum()
```

```python
# Mostra as linhas onde o valor da coluna "Churn" Ã© uma string vazia ou sÃ³ espaÃ§os
dados[dados['Churn'].str.strip() == '']
```

```python
# Mostra as linhas onde o valor da coluna "Charges_Total" Ã© uma string vazia ou sÃ³ espaÃ§os
dados[dados['Charges_Total'].str.strip() == '']
```

```python
#Alterando a coluna Chearges_Total para tipo float64
dados['Charges_Total'] = pd.to_numeric(dados['Charges_Total'], errors='coerce')
print(dados['Charges_Total'].dtype)
```

```python
len(dados)
```

```python
#Excluindo os casos onde a coluna "churn" estÃ¡ vazia
dados = dados[dados['Churn'].str.strip() != '']
```

```python
len(dados)
```

```python
for col in dados.columns:
  print(f'Valores Ãºnicos na coluna "{col}": {dados[col].nunique()}')
  if dados[col].nunique() < 50:
    print(dados[col].unique())
    print('-' *50)
```

```python
#Calculando o gasto diÃ¡rio dos clientes
dados['Daily_cost'] = round(dados['Charges_Total'] /
                            dados['tenure'] / 30, 2)
```

```python
dados[['Charges_Total','tenure','Daily_cost']]
```

```python
#Convertendo os valores 'YES' e 'NO' em 1 e 0
Colunas_binarias = ['Churn', 'Partner','Dependents','PhoneService','MultipleLines','InternetService','OnlineSecurity',
                    'OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling']

for col in Colunas_binarias:
    dados[col] = dados[col].replace({'Yes': 1, 'No': 0}).infer_objects(copy=False)
```

```python
dados.head()
```

```python
for col in dados.columns:
  print(f'Valores Ãºnicos na coluna "{col}": {dados[col].nunique()}')
  if dados[col].nunique() < 50:
    print(dados[col].unique())
    print('-' *50)
```

#ðŸ“Š Carga e anÃ¡lise

```python
cores = ['#5D6D7E', '#6C3483']
```

```python
fig, axs = plt.subplots(1,2, figsize=(9,5))

churn_counts = dados['Churn'].value_counts().sort_index()
cores = sns.color_palette(palette=cores)

bars = axs[0].bar(churn_counts.index.astype(str), churn_counts.values, color=cores)

for i, bar in enumerate(bars):
    bar.set_label(['NÃ£o', 'Sim'][i])

axs[0].set_title('DistribuiÃ§Ã£o do Churn', fontsize=14, weight='bold')
axs[0].set_xlabel('Churn', fontsize=11)
axs[0].set_ylabel('NÃºmero de clientes', fontsize=11)
axs[0].set_xticks([0,1])
axs[0].legend(title='Churn')

for bar in bars:
    yval = bar.get_height()
    axs[0].text(bar.get_x() + bar.get_width()/2, yval + 5, int(yval), ha='center', va='bottom', fontsize=10)


#GrÃ¡fico de Pizza
labels = ['NÃ£o', 'Sim']
axs[1].pie(churn_counts.values, labels=labels,
       autopct='%1.1f%%',
       startangle=90,
       colors=cores,
       explode=(0, 0.05))

axs[1].set_title('ProporÃ§Ã£o de Churn', fontsize=14, weight='bold')

plt.tight_layout()
fig.savefig('churn_distribuicao.png', bbox_inches='tight', dpi=300)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,4))
sns.countplot(data=dados, x='gender', hue='Churn', ax=ax, palette=cores)

ax.set_title('DistribuiÃ§Ã£o de Churn por GÃªnero', fontsize=14, weight='bold')
ax.set_xlabel('GÃªnero', fontsize=11)
ax.set_ylabel('NÃºmero de clientes',fontsize=11)
ax.legend(title='Churn', labels=['NÃ£o', 'Sim'])

for container in ax.containers:
    ax.bar_label(container, fmt='%d', label_type='edge', fontsize=10)

ax.margins(y=0.1)

plt.tight_layout()
fig.savefig('churn_genero.png', bbox_inches='tight', dpi=300)
plt.show()
```

### **4. ConclusÃµes e Insights**

A anÃ¡lise mostrou que o churn na Telecom X estÃ¡ relacionado a alguns padrÃµes consistentes no perfil e comportamento dos clientes. Entre os principais insights, destacam-se:

- **Clientes com contrato mensal** tÃªm uma taxa de churn muito maior do que aqueles com contratos mais longos (anual ou bianual).
- **ServiÃ§os como seguranÃ§a online, backup e suporte tÃ©cnico** estÃ£o associados a menor evasÃ£o. Clientes que nÃ£o contratam esses serviÃ§os tendem a sair com mais frequÃªncia.
- Clientes com **pouco tempo de contrato (tenure baixo)** sÃ£o mais propensos ao churn, o que sugere a importÃ¢ncia dos primeiros meses na fidelizaÃ§Ã£o.
- Clientes que deram churn tendem a ter um **custo mensal mais alto**, o que pode indicar que planos mais caros estÃ£o ligados Ã  evasÃ£o.
- **MÃ©todos de pagamento como `Electronic check` e `Mailed check`** apresentam taxas maiores de churn, enquanto opÃ§Ãµes automÃ¡ticas como **cartÃ£o de crÃ©dito**, **dÃ©bito automÃ¡tico** estÃ£o associadas a menor evasÃ£o.

Esses padrÃµes ajudam a entender melhor o comportamento de saÃ­da dos clientes e servem de base para aÃ§Ãµes mais direcionadas.

### **5. RecomendaÃ§Ãµes**

Com base nos insights obtidos, algumas aÃ§Ãµes podem ser adotadas para reduzir o churn e melhorar a retenÃ§Ã£o de clientes:

- **Incentivar contratos de maior duraÃ§Ã£o**, oferecendo vantagens para quem opta por planos anuais ou bianuais.
- **Promover pacotes com serviÃ§os de valor agregado**, como seguranÃ§a online, backup e suporte tÃ©cnico, que estÃ£o associados a menor evasÃ£o.
- **Criar aÃ§Ãµes especÃ­ficas para novos clientes**, principalmente nos primeiros meses de contrato, como oferecer um acompanhamento mais prÃ³ximo e uma comunicaÃ§Ã£o ativa.
- **Facilitar e incentivar o uso de mÃ©todos de pagamento automÃ¡ticos**, reduzindo o churn financeiro por esquecimento ou dificuldade de pagamento.

Essas medidas, baseadas nos dados analisados, podem ajudar a diminuir a evasÃ£o e melhorar a experiÃªncia geral dos clientes da Telecom X.
